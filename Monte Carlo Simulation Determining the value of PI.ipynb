```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import random
import seaborn as sns
sns.set_style('whitegrid')

np.random.seed(10)
```

```python
# Calculate the value of PI based on number of observation.

n = 1000     # initial number of observations
error_l = []
calculated_pi_list = []
actual_pi = []
iteration = []

for i in range(1, n):
    x = np.random.uniform(low=-1, high = 1, size =[i,1])
    y = np.random.uniform(low=-1, high = 1, size =[i,1])

    inside_circle_condition  = (x**2 + y**2) <= 1
    outside_circle_condition = (x**2 + y**2) > 1
    
    x_inside = x[inside_circle_condition]
    y_inside = y[inside_circle_condition]
    
    calculated_pi = 4*np.sum(inside_circle_condition)/i
    calculated_pi_list.append(calculated_pi)
    actual_pi.append(np.pi)
    error_l.append(np.pi - calculated_pi)
    iteration.append(i)
    
plt.figure(figsize = [8,8])
plt.scatter(x,y,s = 1)
plt.scatter(x_inside, y_inside,color = 'r', s = 1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Monte Carlo simulation, number of observation n = {}  \n'.format(n))
plt.show()

plt.figure(figsize = [10,4])    
plt.scatter(iteration, calculated_pi_list, color = 'r', s = 1, vmin=2, vmax=4, label = 'Simulated pi')
plt.plot(iteration,actual_pi,c = 'b',label = 'Actual pi', linewidth=2.0)
plt.ylim([2,4])
plt.xlabel('Iteration')
plt.ylabel('pi value')
plt.legend(loc="best")
plt.title('Actual/Simulated pi value comparison, number of observation n = {}  \n'.format(n))
plt.show()

plt.figure(figsize = [8,8]) 
sns.distplot(calculated_pi_list,bins=100,color='blue')
plt.title('Calculated pi List with number of observation n = {}'.format(n))
plt.xlim([2.5,3.75])
plt.xlabel('Calculated pi values')
plt.show()

print('Mean:', np.round(np.mean(calculated_pi_list),5), 'Standard deviation:', np.round(np.std(calculated_pi_list),5))
```

Here we can see, with less number of observations, the value of PI is still near original value

```python
df_obs = pd.DataFrame(columns=['Observation', 'mean', 'std', 'RSE', 'actual_pi_value'])
loop_number = 35
n = 1000

for j in range(loop_number):
    error_l = []
    calculated_pi_list = []
    actual_pi = []
    iteration = []

    for i in range(1, n):
        x = np.random.uniform(low=-1, high = 1, size =[i,1])
        y = np.random.uniform(low=-1, high = 1, size =[i,1])

        inside_circle_condition = (x**2 + y**2) <= 1
        outside_circle_condition = (x**2 + y**2) > 1

        x_inside = x[inside_circle_condition]
        y_inside = y[inside_circle_condition]

        calculated_pi = 4*np.sum(inside_circle_condition)/i
        calculated_pi_list.append(calculated_pi)
        actual_pi.append(np.pi)
        error_l.append(np.pi - calculated_pi)
        iteration.append(i)

    rss = 0
    for a in calculated_pi_list:
        rss1 =  np.pi - a
        rss2 = rss1 ** 2
        rss += rss2

    rse        = np.round(np.sqrt(rss)/(len(calculated_pi_list) - 2), 5)
    mean_value = np.round(np.mean(calculated_pi_list),5)
    std_value  = np.round(np.std(calculated_pi_list),5)
    orig_pi    = np.round(np.pi,5)
    
    df_obs     = df_obs.append({'Observation': n, 'mean': mean_value,'std': std_value, 'RSE': rse, 'actual_pi_value' : orig_pi},ignore_index=True)
    
    # print('Observation :', n, 'mean_value :', mean_value, 'std_value :', std_value, 'rse : ', rse)
    
    j += 1
    n += 1000
```

The above figure shows, how the value of pi getting converged with the increasing value of observations.
Residual standard error is getting towards zero with the icreasing number of observations.
